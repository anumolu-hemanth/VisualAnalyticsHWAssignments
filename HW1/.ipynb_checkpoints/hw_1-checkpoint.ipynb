#### 4. Create a new dataframe, df2, that contains only the following fields 'Vehicle_Anti_Theft_Device', 'Driver_Minimum_Age', 'Vehicle_Age_In_Years',  'Annual_Premium', 'Loss_Amount'. Note that 'PolicyNo' is the index of this dataframe as well. To answer the following questions use df2

#### 7. Calculate and print out the Portfolio Premium, Portfolio Loss, and Loss Ratio. Portfolio Premium = sum of all the Annual Premiums of all the policies. Portfolio Loss = sum of all the Loss Amounts of all the policies. Loss Ratio = Portfolio Loss / Portfolio Premium

```python
portfolio_premium = df2['Annual_Premium'].sum()
portfolio_loss = df2['Loss_Amount'].sum()
loss_ratio = portfolio_loss/portfolio_premium
print('Portfolio Premium: ', loss_ratio , '\nPortfolio Loss: ', portfolio_loss , '\nLoss Ratio ',loss_ratio)
```

#### 9. Create a new data frame, df3, that groups the rows by Vehicle_Anti_Theft_Device and aggregates them by taking the mean for all the fields. Note that the index of the new data frame should be Vehicle_Anti_Theft_Device

```python
#copying the 'Vehicle_Anti_Theft_Device', 'Driver_Minimum_Age', 'Vehicle_Age_In_Years',  'Annual_Premium', 'Loss_Amount' fields data from df1
df3 = pd.DataFrame(df1, columns = ['PolicyNo','Vehicle_Anti_Theft_Device','Driver_Minimum_Age','Vehicle_Age_In_Years','Annual_Premium','Loss_Amount'])
#setting policyno as index for df3
df3.set_index('Vehicle_Anti_Theft_Device') 
```

```python
group_by = df3.groupby('Vehicle_Anti_Theft_Device')
df3 = group_by.mean()
df3
```

#### 10. Display the index of the new data frame.

```python
df3.index
```

####  11. Create a new field 'Loss_Ratio' that gives the loss ratio per type of vehicle anti theft device. Loss Ratio = Loss_Amount / Annual_Premium

```python
df3['Loss_Ratio'] = df3['Loss_Amount']/df3['Annual_Premium']
df3
```

#### 12. Sort the data frame by Loss Ratio in ascending order.

```python
df3_sorted = df3.sort_values(by=['Loss_Ratio'])
df3_sorted
```

#### 13. Create a bar plot for 'Vehicle_Anti_Theft_Device' vs 'Loss_Ratio'.

```python
df3['Loss_Ratio'].plot.bar()
```
